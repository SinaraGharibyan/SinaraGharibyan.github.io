```python
import re
import requests
url="https://www.bloomberg.com/quote/SPX:IND"
response = requests.get(url)
page=response.text
output=re.findall('class="price"\s*\S*[.+]?</div>',page)
output1=re.sub(r'<.*?>'," ",str(output))
output2=re.sub(r'class="price">',"Sp500=",output1)
print output2

```
