```python
import urllib2
import re
```

```python
from lxml import html
```

```python
url="https://www.youtube.com/feed/trending"
```

```python
response=urllib2.urlopen(url)
page=response.read()
tree=html.document_fromstring(page)
```

```python
time=tree.xpath('//span[@class="accessible-description"]')
duration=[]
for i in time:
    a=i.text_content()
    c=re.sub('-\sDuration:'," ",a)
    duration.append(c)
    
```

```python
link=tree.xpath('//a[@aria-hidden="true"]')
list_link=[]
for i in link:
    d=i.get('href')
    list_link.append("www.youtube.com"+d)
```

```python
link=tree.xpath('//a[@class="g-hovercard yt-uix-sessionlink       spf-link "]')
list_user=[]
for i in link:
    d=i.get('href')
    list_user.append(d)
    
```

```python
list_views=tree.xpath('//ul[@class="yt-lockup-meta-info"]/li/text()')
list_views_final=[]
list_views_f=[]
for i in list_views:
    m=re.findall("(\d+\S*\d*)\sviews",i)
    list_views_final.append(m)



```

```python
list_views_f=[]
for j in range(len(list_views_final)):
        if len(list_views_final[j])==1:
            x=list_views_final[j][0]
            list_views_f.append(x)

            
```

```python
len(list_views_f)
```

```python
name=tree.xpath('//h3[@class="yt-lockup-title "]/a/@title')

```

```python
data_list=[]
for i in range(0,len(list_user)):
    dictionary={"name":name[i],'url':list_link[i],"duration":duration[i],"username":list_user[i],'views':list_views_f[i]
        
    }
    data_list.append(dictionary)
data_list    
```

```python
import json
with open("youtube_trend_lxml.json","w") as f:
    json.dump(data_list,f,sort_keys = True, indent = 4)
```

```python
import csv
address = "C:\Data_scraping\youtube_trend_lxml.json"
with open(address,"r") as file:
    local_json = json.load(file)
    
with open("from_json_trend_lxml.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["Duration","Name","Url","Username","Views"])
    for item in local_json:
        writer.writerow([item['duration'],item['name'].encode('utf-8'),item['url'],item['username'],item['views']])
```
