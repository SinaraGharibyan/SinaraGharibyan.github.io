```python
import requests
import re
from BeautifulSoup import *
url="https://www.youtube.com/feed/trending"
response=requests.get(url)
page=response.text
soup=BeautifulSoup(page)
list=soup.findAll('ol')
```

```python
dur=list[1].findAll('span',attrs={"class":"accessible-description"})
duration=[]
for i in dur:
    a=i.text
    c=re.sub('-\sDuration:'," ",a)
    duration.append(c)
    
    
```

```python
name=list[1].findAll('h3')
nam=[]
for i in name:
    m=i.text
    n=re.sub('-\sDuration.+'," ",m)
    nam.append(n)
```

```python
list_link=[]
link=list[1].findChildren('a',attrs={'aria-hidden':'true'})
for i in link:
    d=i.get('href')
    list_link.append("www.youtube.com"+d)
```

```python
list_user=[]
link=list[1].findChildren('a',attrs={'class':"g-hovercard yt-uix-sessionlink       spf-link " })
for i in link:
    d=i.get('href')
    list_user.append(d)
```

```python

list_views=[]
link=list[1].findChildren('ul',attrs={'class':"yt-lockup-meta-info" })
for i in link:
    d=i.findAll('li')
    list_views.append(d)
```

```python
list_views_final=[]
for i in list_views:
    m=re.findall("<li>(\d+\S*\d*)\sviews",str(i))
    list_views_final.append(m)
   
```

```python
for i in range(len(list_views_final)):
    list_views_final[i]=list_views_final[i][0]

             
```

```python
data_list=[]
for i in range(0,len(list_user)):
    dictionary={"name":nam[i],'url':list_link[i],"duration":duration[i],"username":list_user[i],'views':list_views_final[i]
        
    }
    data_list.append(dictionary)
data_list    
```

```python
import json
with open("youtube_trend.json","w") as f:
    json.dump(data_list,f,sort_keys = True, indent = 4)
```

```python
import csv
address = "C:\Data_scraping\youtube_trend.json"
with open(address,"r") as file:
    local_json = json.load(file)
    
with open("from_json_trend.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["Duration","Name","Url","Username","Views"])
    for item in local_json:
        writer.writerow([item['duration'],item['name'].encode('utf-8'),item['url'],item['username'],item['views']])
```
